# Task 1: Exploring and Visualizing a Simple Dataset

```python
# Dataset: Iris

# Importing required libraries
import pandas as pd                
import seaborn as sns               
import matplotlib.pyplot as plt     

%matplotlib inline
```

## Loading The Dataset

```python
# 1. Load the Iris dataset
iris = sns.load_dataset("iris")

# Display the first few rows to understand the structure of the dataset
print("First 5 rows of the dataset:")
print(iris.head())
```

## Basic Dataset Information

```python
# Print the shape (rows, columns)
print("\nShape of dataset:", iris.shape)

# Display all column names
print("\nColumn names:", iris.columns.tolist())

# Show detailed information about dataset (data types, missing values, etc.)
print("\nDataset Info:")
print(iris.info())

# Summary statistics for numerical features
print("\nSummary Statistics:")
print(iris.describe())
```

## Data Visualisation

```python
# Scatter Plot 
# Shows the relationship between two features, colored by species
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="sepal_width",
    hue="species"  
)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()


# Histograms 
# Helps observe the distribution of each numeric feature
iris.hist(figsize=(10, 8))
plt.suptitle("Histograms of Iris Dataset Features")
plt.show()


# -------- Box Plots --------
# Useful to detect outliers and understand spread of data
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Dataset Features")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```
